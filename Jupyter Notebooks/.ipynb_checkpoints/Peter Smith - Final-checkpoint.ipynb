```python
# Import code from source [1]

from __future__ import absolute_import, division, print_function, unicode_literals

%load_ext tensorboard

# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras


# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

### End of source [1]

# additional utilities imported as they were used for experimentation.
import datetime
import os
import copy
```

```python
# Neural Network training has been split into component functions. 
# This allows for component testing, component experimentation and neater code.
# It also allows for less repeated, redundant code.

# Global variables declared, so that functions can access data.
# They are initiallised as None, and will be assigned values later.
train_labels = None
train_images = None
test_labels = None
test_images = None
class_names = None
history = None
```

```python
# This functions downloads the CIFAR-10 dataset and assigns the data to the global variables.
def processDataset():
    # This is necessary so the data is not written to local variables, 
    # as this would lead to other functions not being able to access the data.
    global train_labels
    global train_images
    global test_labels
    global test_images
    global class_names
    
    # Download and preparation code from source [1]
    
    # Load the data onto the variables
    (train_images, train_labels), (test_images, test_labels) = keras.datasets.cifar10.load_data()
    
    # The train_labels of the dataset are represented by numbers.
    # To make this more readable and usable, we use class_names to give meaning to those numbers.
    class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
    
    # We standardise the values so that are between 0 and 1.
    train_images = train_images / 255.0
    test_images = test_images / 255.0
    
    ### End of source [1]

# I process the dataset here initially as component testing to make sure everything so far has been imported correctly, 
# and it also downloads the data, readying it for use elsewhere. 
print(train_labels)
processDataset()
print(train_labels)
```

```python
# This function takes a subset of images from the dataset and displays them alongside their label.
# This demonstrates the type of images we're classifying (e.g. colour images) and displays them.
def showDataset():
    
    # matplotlib code from source [1]
    
    plt.figure(figsize=(10,10))
    for i in range(25):
        plt.subplot(5,5,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(train_images[i], cmap=plt.cm.binary)
        plt.xlabel(class_names[train_labels[i][0]])
    plt.show()
    
    ### End of source [1]
    
showDataset()
```

```python
# This function takes a trained model, runs the test dataset against it in order to 
# find the final accuracy and loss.
def evaluateModel(model):
    
    # Evaluation code from source [1]
     
    # This runs model.evaluate to test the trained model with test_images and test_labels
    # to get the loss (test_loss) and accuracy (test_acc)
    test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
    
    ### End of source [1]
    
    # I then print these values again. 
    # This isn't strictly necessary given the model.evaluate attribute 'verbose=2' will show these values as well.
    print('Acc: ' + str(test_acc))
    print('Loss:' + str(test_loss))
```

```python
# This function will take a trained model and print a subset of the testing dataset
# alongside a bar graph showing the neural networks predictions on how to classify that image.
def showPredictions(model):
    
    # Predictions code from source [2]
    
    # We create a probability model by adding a softmax layer at the end of our trained model. 
    probability_model = keras.Sequential([model, keras.layers.Softmax()])
    
    # We then run the .predict function using the test_images dataset.
    predictions = probability_model.predict(test_images)
    
    # We initialise values for our graph.
    num_rows = 5
    num_cols = 3
    num_images = num_rows*num_cols
    
    # We use two subplots. For one, we run the plot_image function, and the other we run the plot_value_array.
    plt.figure(figsize=(2*2*num_cols, 2*num_rows))
    for i in range(num_images):
        plt.subplot(num_rows, 2*num_cols, 2*i+1)
        plot_image(i, predictions[i], test_labels, test_images)
        plt.subplot(num_rows, 2*num_cols, 2*i+2)
        plot_value_array(i, predictions[i], test_labels)
    plt.tight_layout()
    plt.show()

# This function plots the specified image in the subplot.
def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array, true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap=plt.cm.binary)

    # If the prediction is correct, it will be blue, otherwise it will be red.
    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label[0]]),
                                color=color)

# This function plots the specified value array in the subplot.
def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array, true_label[i]
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label[0]].set_color('blue')

    ### End of source [2]
```

```python
# This function plots the training and validation accuracy of a model throughout training against the epoch.
def showAccuracy():
    plt.figure(1)
    
    # matplotlib code as from source [1]
    
    plt.plot(history.history['accuracy'], label='accuracy')
    plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.ylim([0, 1])
    plt.legend(loc='lower right')
    
    ### End of source [1]
```

```python
# This function plots the training and validation loss of a model throughout training against the epoch. 
def showLoss():
    plt.figure(2)
    
    # matplotlib code adapted from source [1]
    
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label = 'val_loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.ylim([0, 2])
    plt.legend(loc='upper right')
    
    ### End of source [1]
```

```python
# This function is the master function and trains the model and runs the evaluation and shows the results. 
# It takes a model, an integer value and a string.
def trainModel(model, epochCount, identifier):
    global history
    
    # Prints a summary of the model. This shows the architecture and displays the number of parameters at each stage.
    model.summary()
    
    # Tensorboard code as from source [3]
    
    # Use the identifier parameter to add to the log path. This will make it easier to work with in Tensorboard.
    log_dir = os.path.join(
    "logs",
    "final",
    datetime.datetime.now().strftime("%Y%m%d-%H%M%S") + identifier,
    )
    
    tensorboard_callback = keras.callbacks.TensorBoard(log_dir=log_dir,histogram_freq=1)
    
    %tensorboard --logdir logs/final
    
    # Tensorflow code as from source [1]
    
    # Compile the model, 
    # using the adam optimiser, a very common optimiser with a decaying learning rate 
    # using the sparse categorical cross entropy, a very common loss function for image classification
    model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
    
    # Train the model with train_images and train_labels, writing the metrics to the history variable.
    # Taking the epochCount parameter and training the model for that many epochs. This parameter allows easy experimentation.
    # After each epoch, this function will test the model with the 'validation_data' (test_images and test_labels).
    # The 'callbacks' also output the metrics to Tensorboard as well as history.
    history = model.fit(train_images, train_labels, epochs=epochCount, 
                    validation_data=(test_images, test_labels), callbacks=[tensorboard_callback])
    
    ### End of source [1]
    
    ### End of source [3]
    
    # After training our model, we evaluate it to get a final accuracy and loss.
    # We then show some predictions on the test-set.
    # We then plot the accuracy and loss graphs.
    evaluateModel(model)
    showPredictions(model)
    showAccuracy()
    showLoss()
```

# Model Training
### Model 1, using ReLU

```python
# I make each model again for each test. This may not be the most efficient. 
# However when I tried making the models global variables, and assigning it to a local variable, which I then trained,
# The global variable was also trained as I was assigning the local variable a reference to the existing object, 
# not creating a new object. 
# There is likely a fix for this, but I did not have the time to investigate, and this method works well.

# Model code from source [1]

# I add the layers differently than in the tutorial. Rather than model.add(layer), I am passing an array of layers.
# Beyond this, the parameters remain largely unchanges.
testModel1 = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), input_shape=(32, 32, 3)),
    keras.layers.Activation('relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3)),
    keras.layers.Activation('relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3)),
    keras.layers.Activation('relu'),
    keras.layers.Flatten(),
    keras.layers.Dense(64),
    keras.layers.Activation('relu'),
    keras.layers.Dense(10)
])

### End of source [1]

# I train each model for 50 epochs.
trainModel(testModel1, 50, " Model 1 ReLU")
```

### Model 1, using TanH

```python
# This model is the same as the one before, only every 'relu' has been replaced by 'tanh'.
# I use the same variable, testModel1, but as I am reassigning the value to a new object, 
# there's no chance of the previous run affecting this one.

# Model code from source [1]

testModel1 = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), input_shape=(32, 32, 3)),
    keras.layers.Activation('tanh'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3)),
    keras.layers.Activation('tanh'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3)),
    keras.layers.Activation('tanh'),
    keras.layers.Flatten(),
    keras.layers.Dense(64),
    keras.layers.Activation('tanh'),
    keras.layers.Dense(10)
])

### End of source [1]

#Again, I train the model for 50 epochs.
trainModel(testModel1, 50, " Model 1 TanH")
```

### Model 2, using ReLU

```python
# Model code from source [1]

# Dropout placement inspired by sources [4] and [5]

testModel2 = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), input_shape=(32, 32, 3)),
    keras.layers.Activation('relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Dropout(0.25),
    keras.layers.Conv2D(64, (3, 3)),
    keras.layers.Activation('relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Dropout(0.25),
    keras.layers.Conv2D(64, (3, 3)),
    keras.layers.Activation('relu'),
    keras.layers.Flatten(),
    keras.layers.Dense(64),
    keras.layers.Activation('relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(10)
])

### End of source [4] and [5]

### End of source [1]

trainModel(testModel2, 50, " Model 2 ReLU")
```

### Model 2, using TanH

```python
# Model code from source [1]

# Dropout placement inspired by sources [4] and [5]

testModel2 = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), input_shape=(32, 32, 3)),
    keras.layers.Activation('tanh'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Dropout(0.25),
    keras.layers.Conv2D(64, (3, 3)),
    keras.layers.Activation('tanh'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Dropout(0.25),
    keras.layers.Conv2D(64, (3, 3)),
    keras.layers.Activation('tanh'),
    keras.layers.Flatten(),
    keras.layers.Dense(64),
    keras.layers.Activation('tanh'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(10)
])

### End of source [4] and [5]

### End of source [1]

trainModel(testModel2, 50, " Model 2 TanH")
```
